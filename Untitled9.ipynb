```python
import pandas as pd
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
start = dt.datetime(2020, 7, 7)
end = dt.datetime.now()
```

```python
tickers = ['GOOGL', 'RWE.DE', 'O', 'KO', 'CRWD', 'LIN', 'CLH', 'ENPH', 'PEP', 'MSFT', 'ASML.AS', 'PRX.AS', 'NPI.TO', 'PG','EL','VALE','ALB','MP', 'BHP.AX']
```

```python
prices_df = yf.download(' '.join(tickers), start, end) ['Adj Close']
prices_df
```

```python
stock_returns = prices_df.pct_change()
```

```python
cumulative_returns = (1 + stock_returns).cumprod()
cumulative_returns
```

```python
weights = [0.05,0.05,0.05,0.05,0.05,
0.05,0.05,0.05,0.05,0.05,
0.05,0.05,0.05,0.05,0.05,
0.05,0.05,0.05,0.05]
weights2 = [0.01,0.01, 0.01, 0.37,0.01,
 0.01,0.01,0.01, 0.01, 0.09,
0.01,0.09,0.01, 0.01, 0.13,
0.05,0.13,0.01,0.01]

weights3 = [0.01251203,  0.08373436,  0.06544755,  0.186718  ,  0.10298364,
       -0.06352262,  0.07314726,  0.08950914,  0.07892204,  0.10105871,
        0.02694899,  0.09143407, -0.02694899,  0.03368624,  0.04716073,
        0.05967276, -0.01251203,  0.00866218,  0.04138595]
```

```python
port1 = np.dot (cumulative_returns.mean(), weights)
port2 = np.dot (cumulative_returns.mean(), weights2)
port3 = np.dot (cumulative_returns.mean(), weights3)
```

```python
porta = np.dot (cumulative_returns, weights)
portb = np.dot (cumulative_returns, weights2)
portc = np.dot (cumulative_returns, weights3)
```

```python
dates = cumulative_returns.index
dict_port_dates = {'Date' :dates, 'TotalReturnA' : porta, 'TotalReturnB' :portb, 'TotalReturnC':portc}
dict_port_dates
df = pd.DataFrame(dict_port_dates)
df = df.set_index('Date')
```

```python
df['TotalReturnA'].plot(label="Weights 1 Portfolio", figsize=(15,7))
df['TotalReturnB'].plot(label='Weights 2 Portfolio', figsize=(15,7))
df['TotalReturnC'].plot(label='Weights 3 Portfolio', figsize=(15,7))
plt.legend()
```

```python
var = stock_returns.var()
var = (var **(12/39))
return_year = ((1+ stock_returns).prod())**(12/39) - 1
comparing = pd.DataFrame(var, columns=['Variance'])
y_return = pd.DataFrame(return_year, columns=['ReturnY'])
optimizer_df = pd.concat([comparing['Variance'],y_return['ReturnY']],axis=1)
optimizer_df['StDev'] = optimizer_df['Variance']**(1/2) # **0.5
optimizer_df['RiskRewardRatio'] = (optimizer_df['ReturnY'] / optimizer_df['StDev']).round(2)
# A stock's historical variance measures the difference between the stock's returns for different periods and its average return.
# A stock with a lower variance typically generates returns that are closer to its average.
optimizer_df
```

```python
allocation = optimizer_df
allocation['AllocationMaxRiskrwShortIncl'] = (allocation['RiskRewardRatio'])
sum_rw = np.sum(allocation.AllocationMaxRiskrwShortIncl)
result = (1 / sum_rw)
allocation['AllocationMaxRiskrwShortIncl'] = (allocation['AllocationMaxRiskrwShortIncl'] * result)
MaxReturnShortincl = (allocation.AllocationMaxRiskrwShortIncl.to_numpy())
MaxReturnShortincl
#allocation
```

```python
allocation = optimizer_df
allocation
allocation['RiskRewardNoShort'] = allocation['RiskRewardRatio']
allocation.RiskRewardNoShort = np.where(allocation.RiskRewardNoShort < 0, 0.001, allocation. RiskRewardNoShort)
sum_ns = np.sum(allocation.RiskRewardNoShort)
allocation['AllocationMaxRiskrwNoShort'] = (allocation['RiskRewardNoShort'] / sum_ns )
MaxReturnNoShort = (allocation.AllocationMaxRiskrwNoShort.to_numpy().round(5))
MaxReturnNoShort
#allocation
```

```python
sum_var = np.sum(allocation.Variance)
allocation['AllocationMinVar'] = (sum_var /allocation['Variance'] )
res = np.sum(allocation.AllocationMinVar)
allocation['AllocationMinVar'] = (sum_var /allocation['Variance'] ) / res
MinVararray = (allocation.AllocationMinVar.to_numpy().round(5))
MinVararray
#allocation
```
